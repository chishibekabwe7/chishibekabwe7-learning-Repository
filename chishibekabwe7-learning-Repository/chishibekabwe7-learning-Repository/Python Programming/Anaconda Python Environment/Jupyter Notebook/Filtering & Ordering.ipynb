```python
#Python Pandas loc Tutorial
df3 = pd.read_csv(r"C:\Users\Chishibe Kabwe\Documents\Data Cleaning Samples\W3Schools Data Cleaning Sample.csv")
df3
```

```python
#Filtering an entire row.
out_put1 = df3.loc[1]
out_put1
```

```python
#Slicing rows with the loc statement.
out_put2 = df3.loc[1:11]
out_put2 
```

```python
#Extracting specific columns from our dataset.
Duration_Calories_Column = df3.loc[:,['Duration','Calories']]
Duration_Calories_Column
```

```python
#Extracting Specific columns from a sliced dataset.
Calories_Duration_Column = df3.loc[22:30,['Duration','Calories']]
Calories_Duration_Column
```

```python
#Filtering a specific element by specifing the row and column with the loc statement.
Specific_Pulse = df3.loc[8,'Pulse']
Specific_Pulse
```

```python
#Specifing our output with boolean logic.
Pulse_with_limit = df3.loc[df3['Pulse'] > 100]
Pulse_with_limit 
```

```python
#Python Pandas iloc Tutorial
print(df3)
```

```python
#Filtering an entire row with iloc (indexing)
row_0 = df3.iloc[0]
print(row_0)
```

```python
#Filtering multiple rows by slicing with indexing
sliced_output  = df3.iloc[4:10]
print(sliced_output)
```

```python
#Filtering specific columns with indexing
specific_columns_3_4 = df3.iloc[: ,[3,4]]
print(specific_columns_3_4)
```

```python
#Filtering sliced specific columns with indexing
specific_columns_3_4 = df3.iloc[2:3, 3:4]
print(specific_columns_3_4)
```

```python
#Filtering a specific element by specifing the index row and index column of an element.
#Eg filter out 120 from df3
Element = df3.iloc[11,3]
print(Element)
```

```python
#Note you can also use negative indexes as -1 starts from the end 0 starts from the first element.
```

```python
#Python Pandas Ordering Tutorial
print(df3)
```

```python
#Sorting a specific column
sorted_by_Pulse = df3.sort_values(by='Pulse',ascending=False)
print(sorted_by_Pulse)
```

```python
#Sorting Multiple columns
sorted_columns = df3.sort_values(by = ['Date','Pulse','Calories'])
print(sorted_columns)
```
